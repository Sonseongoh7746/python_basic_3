```python
# 1. 데이터프레임 탐색
weather = pd.read_csv("weather.csv", index_col = 0)
weather
```

```python
# 2. 데이터 정제

# 1) 개별적인 열의 개수 확인 
print(weather.count())

# 2) 결측치 탐지 및 처리
missing_data = weather[weather['평균 풍속'].isna()]
print(missing_data )

#3) 결측 값을 평균 풍속으로 채움, inplace를 True로 설정해 원본 데이터를 수정
mean_wind = weather['평균 풍속'].mean()
weather.fillna(mean_wind, inplace=True)
weather
```

```python
# 3.[데이터분석] "평균 기온" 열에서 어느 연도에  평균 기온이 가장 높았는지 연도별 평균 기온을 막대그래프로 그리기
```

```python
# 3.1 데이터프레임 평탄화
weather.head()
```

```python
weather.reset_index(inplace=True)
weather.head()
```

```python
# 3.2 weather['일시']의 연도 값을 추출하여 weather['year']에 저장하여 데이터프레임에 추가하기

weather['year'] = pd.DatetimeIndex(weather['일시']).year

# 3.3 'year' 별 평균 기온의 평균값을 집계하기

yearly_temp = weather.groupby('year')['평균 기온'].mean()

# 3.4 막대 그래프로 그리기
years = np.arange(2010,2021)
plt.bar(years, yearly_temp, color="green")
plt.xlabel('연도')
plt.ylabel('평균 기온')
plt.title("연도별 평균 기온")
plt.show()
```

```python
weather.head()
```

```python
# [데이터분석] 4. 어느 월에  평균 기온이 가장 높았는지 월별 평균 기온을 막대그래프로 그리기
# month 속성을 사용하여 달별로 구분하여 평균 기온 값을 담기
weather['month'] = pd.DatetimeIndex(weather['일시']).month

monthly_temp = weather.groupby('month')['평균 기온'].mean()

months = np.arange(1,13)
plt.bar(months, monthly_temp, color="green")
plt.xlabel('월')
plt.ylabel('평균 기온')
plt.title("월별 평균 기온")
plt.show()
```

```python
# [데이터 분석] 5. 기상 정보를 월별로 막대그래프로 그려서 분석하기
monthly_means = weather.groupby('month')[['평균 기온','최대 풍속', '평균 풍속']].mean()
monthly_means

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].bar(np.arange(1,13), monthly_means['평균 기온'], color = "red")
axs[0].set_title("월별 평균 기온")
axs[0].set_xlabel("월별")
axs[0].set_ylabel("평균 기온")


axs[1].bar(np.arange(1,13), monthly_means['최대 풍속'], color = "red")
axs[1].set_title("월별 최대 풍속")
axs[1].set_xlabel("월별")
axs[1].set_ylabel("최대 풍속")


axs[2].bar(np.arange(1,13), monthly_means['평균 풍속'], color = "blue")
axs[2].set_title("월별 최대 풍속")
axs[2].set_xlabel("월별")
axs[2].set_ylabel("평균 풍속")
```

```python
# [데이터분석] 6. 기상 정보를 연도별로 막대그래프로 그려서 분석하기
yearly_means = weather.groupby('year')[['평균 기온','최대 풍속', '평균 풍속']].mean()

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].bar(np.arange(2010,2021), yearly_means['평균 기온'], color = "red")
axs[0].set_title("연도별 평균 기온")
axs[0].set_xlabel("연도")
axs[0].set_ylabel("평균 기온")


axs[1].bar(np.arange(2010,2021),yearly_means['최대 풍속'], color = "red")
axs[1].set_title("연도별 최대 풍속")
axs[1].set_xlabel("연도")
axs[1].set_ylabel("최대 풍속")


axs[2].bar(np.arange(2010,2021),yearly_means['평균 풍속'], color = "blue")
axs[2].set_title("연도별 최대 풍속")
axs[2].set_xlabel("연도")
axs[2].set_ylabel("평균 풍속")
```

## 5) 연도별 "평균 기온"의 평균에 대한 추이를 선 그래프로 그리시오.

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('weather.csv')

# '일시' 열을 날짜 형식으로 변환
df['일시'] = pd.to_datetime(df['일시'])

# '연도' 열 추가
df['연도'] = df['일시'].dt.year

# 연도별 "평균 기온"의 평균 계산
mean_temp_by_year = df.groupby('연도')['평균 기온'].mean()

# 선 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(mean_temp_by_year.index, mean_temp_by_year.values, marker='o', linestyle='-')
plt.title('연도별 평균 기온 추이')
plt.xlabel('연도')
plt.ylabel('평균 기온')
plt.grid(True)
plt.xticks(mean_temp_by_year.index)
plt.tight_layout()
plt.show()
```

## 6) 연도별 "최대 풍속"의 평균을 막대 그래프로 그리시오.

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('weather.csv')

# '일시' 열을 연도로 변환
df['연도'] = pd.to_datetime(df['일시']).dt.year

# 연도별 "최대 풍속"의 평균 계산
mean_max_wind_speed_by_year = df.groupby('연도')['최대 풍속'].mean()

# 막대 그래프 그리기
plt.figure(figsize=(10, 6))
mean_max_wind_speed_by_year.plot(kind='bar', color='skyblue')
plt.title('연도별 평균 최대 풍속')
plt.xlabel('연도')
plt.ylabel('평균 최대 풍속')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 7) 연도별 "평균 풍속"의 합을 히스토그램 그래프로 그리시오


```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('weather.csv')

# '일시' 열을 연도로 변환
df['연도'] = pd.to_datetime(df['일시']).dt.year

# 연도별 "평균 풍속"의 합 계산
sum_avg_wind_speed_by_year = df.groupby('연도')['평균 풍속'].sum()

# 히스토그램 그래프 그리기
plt.figure(figsize=(10, 6))
plt.hist(sum_avg_wind_speed_by_year, bins=10, color='skyblue', edgecolor='black')
plt.title('연도별 평균 풍속 합의 분포')
plt.xlabel('평균 풍속 합')
plt.ylabel('빈도')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 8) 2015년과 2016년의 "평균 기온"의 평균을 비교하여 어느 해가 더 더웠는지 출력하시오.

```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('weather.csv')

# '일시' 열을 연도로 변환
df['연도'] = pd.to_datetime(df['일시']).dt.year

# 2015년과 2016년의 "평균 기온"의 평균 계산
mean_temp_2015 = df[df['연도'] == 2015]['평균 기온'].mean()
mean_temp_2016 = df[df['연도'] == 2016]['평균 기온'].mean()

# 결과 출력
if mean_temp_2015 > mean_temp_2016:
    print("2015년이 더 더웠습니다.")
elif mean_temp_2015 < mean_temp_2016:
    print("2016년이 더 더웠습니다.")
else:
    print("2015년과 2016년의 평균 기온이 동일합니다.")
```

## 9) 월별 "평균 기온" 의 합이 가장 적은 달과 가장 많은 달을 출력하시오.

```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('weather.csv')

# '일시' 열을 날짜 형식으로 변환하여 '월' 열 추가
df['월'] = pd.to_datetime(df['일시']).dt.month

# 월별 "평균 기온"의 합 계산
mean_temp_sum_by_month = df.groupby('월')['평균 기온'].sum()

# 가장 적은 평균 기온 합을 가진 월 출력
min_mean_temp_month = mean_temp_sum_by_month.idxmin()
min_mean_temp_sum = mean_temp_sum_by_month.min()
print(f"평균 기온 합이 가장 적은 월: {min_mean_temp_month}, 합: {min_mean_temp_sum}")

# 가장 많은 평균 기온 합을 가진 월 출력
max_mean_temp_month = mean_temp_sum_by_month.idxmax()
max_mean_temp_sum = mean_temp_sum_by_month.max()
print(f"평균 기온 합이 가장 많은 월: {max_mean_temp_month}, 합: {max_mean_temp_sum}")
```

## 10) 연도별 "최대 풍속"의 평균이 가장 큰 연도을 출력하시오.

```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('weather.csv')

# '일시' 열에서 연도 정보 추출
df['연도'] = pd.to_datetime(df['일시']).dt.year

# 연도별 '최대 풍속'의 평균 계산
mean_max_wind_speed_by_year = df.groupby('연도')['최대 풍속'].mean()

# 평균이 가장 큰 연도 추출
max_avg_wind_speed_year = mean_max_wind_speed_by_year.idxmax()

print(f"연도별 '최대 풍속'의 평균이 가장 큰 연도: {max_avg_wind_speed_year}")
```
